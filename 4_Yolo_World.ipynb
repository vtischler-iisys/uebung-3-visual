```python
from ultralytics import YOLOWorld
import matplotlib.pyplot as plt
import cv2

model = YOLOWorld("yolov8s-worldv2.pt")
```

```python
class_list = ["tie", "hat", "flower"]
model.set_classes(class_list)
```

```python
def display_saved_image(file_path):
    img = cv2.imread(file_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.show()
```

```python
results = model.predict("../VLM-Try_around//images/Hase.png", max_det=100, iou=0.01, conf=0.01)

results[0].save(filename='result.jpg')
display_saved_image('result.jpg')
```
