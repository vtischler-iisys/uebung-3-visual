## Initialize the model

```python
import torch

model = torch.hub.load("ultralytics/yolov5", "yolov5s")  
```

```python
# Inference on images
img = "https://ultralytics.com/images/zidane.jpg"  # Can be a file, Path, PIL, OpenCV, numpy, or list of images

# Run inference
results = model(img)

# Display results
results.print()
results.show()
```
